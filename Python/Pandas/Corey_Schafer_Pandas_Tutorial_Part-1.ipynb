### Lec-4: Filtering - Using conditionals to Filter Rows and Columns.

```python
people = {
    "first": ["Subhrajit", "Rahul", "Rohit"],
    "last": ["Guchait", "N", "N"],
    "email": ["subhrajitguchait@email.com", "rahul@email.com", "rohit@email.com"]
}
```

```python
df2 = pd.DataFrame(people)
```

```python
df2
```

```python
df2.set_index('email', inplace=True)
```

```python
df2
```

```python
df2['last'] == 'N'
```

```python
#Assigning a variable to any filter is a very good practice.
filt = (df2['last'] == 'N')
```

```python
type(df2['last'] == 'N')
```

```python
df2[filt]
```

```python
df2.loc[filt]
```

```python
type(df2[filt])
```

```python
#if we want the last name column:
df2.loc[filt, 'last']
```

```python
df2.reset_index(inplace=True)
```

```python
df2
```

```python
filt2 = ((df2['last'] == 'N') | (df2['first'] == 'Rohit'))
```

```python
df2.loc[filt2, 'email']
```

```python
# Using a '~' against a filter will negate the filter
# (i.e. will output the complete opposite of any given filter)
# very handy while performing EDA.
df2.loc[~filt2, 'email']
```
